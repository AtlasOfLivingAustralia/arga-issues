## Austraits - analysis of traits vs taxa coverage

```python
import pandas as pd

df = pd.read_csv("../data/traits.csv", dtype='object', parse_dates=['collection_date']) 
df.info()
```

```python
df.head(20)
```

```python
def display_full(x):
    with pd.option_context('display.max_rows', None,
                           'display.max_columns', None,
                           'display.width', 2000,
                           'display.float_format', '{:20,.2f}'.format,
                           'display.max_colwidth', None):
        display(x)
```

```python
display_full(df.groupby(['trait_name'])['taxon_name'].count().reset_index(name='count').sort_values(['count'], ascending=False).reset_index())
```
